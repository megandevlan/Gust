```python
# Grabbed from Brian M. to use time midpoints, not end periods
def cesm_correct_time(ds):
    """Given a Dataset, check for time_bnds,
       and use avg(time_bnds) to replace the time coordinate.
       Purpose is to center the timestamp on the averaging inverval.   
       NOTE: ds should have been loaded using `decode_times=False`
    """
    assert 'time_bnds' in ds
    assert 'time' in ds
    correct_time_values = ds['time_bnds'].mean(dim='nbnd')
    # copy any metadata:
    correct_time_values.attrs = ds['time'].attrs
    ds = ds.assign_coords({"time": correct_time_values})
    ds = xr.decode_cf(ds)  # decode to datetime objects
    return ds
```

```python
def preprocess_h0(ds):
    
    # ds         = cesm_correct_time(ds)
    # ds['time'] = ds.indexes['time'].to_datetimeindex()     
    
    ## Add a few shorthand variables 
    ds['windSpeedBot'] = (('time','lat','lon'),    
                                    np.sqrt(ds['U'].isel(lev=-1).values**2 + ds['V'].isel(lev=-1).values**2) )
    ds['windSpeedBot'].attrs['units']   = "m/s"
        
    return ds
```

```python
def preprocess_h1(ds):
    
    # ds         = cesm_correct_time(ds)
    # ds['time'] = ds.indexes['time'].to_datetimeindex()     
    
    ## Add a few shorthand variables 
    ds['windSpeedBot'] = (('time','lat','lon'),    
                                    np.sqrt(ds['U'].isel(lev=-1).values**2 + ds['V'].isel(lev=-1).values**2) )
    ds['windSpeedBot'].attrs['units']   = "m/s"
        
    return ds
```

```python
def meanTimeAndRegion(DS,monStart,monEnd,yrMin,yrMax,
                     minLat,maxLat,minLon,maxLon): 
    
    iTime = np.where( (DS['time.month']>=monStart) & (DS['time.month']<=monEnd) & (DS['time.year']>=yrMin) & (DS['time.year']<=yrMax) )[0]
    iLat  = np.where( (DS['lat']>=minLat) & (DS['lat']<=maxLat) )[0]
    iLon  = np.where( (DS['lon']>=minLon) & (DS['lon']<=maxLon) )[0]

    selDS     = DS.isel(time=iTime, lat=iLat, lon=iLon).mean(dim='time')
    
    
    return selDS
```

```python
## Say we want to look into just JJA in the NW Pacific...

monStart = 6
monEnd   = 8
yrMin      = 2
yrMax      = 5

# ## Tropical NW Pacific
# minLat     = 5
# maxLat     = 30
# minLon     = 90
# maxLon     = 145   # Original domain
# # maxLon     = 170 # 3/30: expand? 

## ~global
minLat     = -80
maxLat     = 80
minLon     = 0
maxLon     = 359   


selUA_58L = meanTimeAndRegion(UA_h1,monStart,monEnd,yrMin,yrMax,minLat,maxLat,minLon,maxLon)
selLP_58L = meanTimeAndRegion(LP_h1,monStart,monEnd,yrMin,yrMax,minLat,maxLat,minLon,maxLon)
selUA_32L = meanTimeAndRegion(UA32_h1,monStart,monEnd,yrMin,yrMax,minLat,maxLat,minLon,maxLon)
selLP_32L = meanTimeAndRegion(LP32_h1,monStart,monEnd,yrMin,yrMax,minLat,maxLat,minLon,maxLon)

```

```python
## Regrid model data to ERAI resolution?
## Using the regridding approach ADF uses: 
##  https://github.com/NCAR/ADF/blob/7055be4ad2cbd16e37d397ad3d7ed5c1d28b42e1/scripts/regridding/regrid_and_vert_interp.py

def regridToModel(camDS, obsDS, camVarName, obsVarName):
    tothis   = obsDS[obsVarName]
    fromthis = camDS[camVarName]

    ## Option 2, not used in ADF right now: 
#     newlat = tothis['lat']
#     newlon = tothis['lon']
#     coords = dict(fromthis.coords)
#     coords['lat'] = newlat
#     coords['lon'] = newlon

#     test = fromthis.interp(coords)

    regridData = fromthis.interp_like(tothis)
    
    return regridData

```

```python
def regrid_data(fromthis, tothis, method=1):
    """Regrid data using various different methods"""

    #Import necessary modules:
    import xarray as xr

    if method == 1:
        # kludgy: spatial regridding only, seems like can't automatically deal with time
        if 'time' in fromthis.coords:
            result = [fromthis.isel(time=t).interp_like(tothis) for t,time in enumerate(fromthis['time'])]
            result = xr.concat(result, 'time')
            return result
        else:
            return fromthis.interp_like(tothis)
    elif method == 2:
        newlat = tothis['lat']
        newlon = tothis['lon']
        coords = dict(fromthis.coords)
        coords['lat'] = newlat
        coords['lon'] = newlon
        return fromthis.interp(coords)
    elif method == 3:
        newlat = tothis['lat']
        newlon = tothis['lon']
        ds_out = xr.Dataset({'lat': newlat, 'lon': newlon})
        regridder = xe.Regridder(fromthis, ds_out, 'bilinear')
        return regridder(fromthis)
    elif method==4:
        # geocat
        newlat = tothis['lat']
        newlon = tothis['lon']
        result = geocat.comp.linint2(fromthis, newlon, newlat, False)
        result.name = fromthis.name
        
        return result
```

```python
## Get obs grid 
obsGrid = erai_jja.WIND_MAG_SURF.isel(time=0).squeeze()

## Now regrid the data
regrid_selLP_JJA    = regrid_data(selLP_JJA, obsGrid, method=1)
regrid_selLP32_JJA  = regrid_data(selLP32_JJA, obsGrid, method=1)

```

```python
def plotBiases(camDS, obsDS, camVarName, obsVarName, 
               obsMultiplier, unitString, titleString): 
    # -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- 
    ## Regrid model data to OBS resolution
    
    regridModel = regridToModel(camDS, obsDS, camVarName, obsVarName)
    obsModel    = obsDS[obsVarName]
    
    # -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- 
    ## Plot bias map
    fig,axs = plt.subplots(1,1, figsize=(10,7),
                          subplot_kw={'projection':ccrs.Robinson(central_longitude=180)})

     ## Plot difference 
    diffVar = np.squeeze(regridModel) - np.squeeze(obsModel*obsMultiplier)

    cMax      = np.nanpercentile(diffVar, 99) 
    cMin      = np.nanpercentile(diffVar, 1)
    evenPoint = np.nanmax([abs(cMax),abs(cMin)]) 
    cStep     = (evenPoint*2)/100.0

    contour_plot = axs.contourf(regridModel.lon.values, regridModel.lat.values, 
                    diffVar, 
                    np.arange(-evenPoint, evenPoint+cStep, cStep), 
                    cmap='bwr',extend='both',
                    transform=ccrs.PlateCarree())

    # Add colorbar
    ax_position = axs.get_position()
    cbar_ax = fig.add_axes([ax_position.x0, ax_position.y0-0.05, ax_position.width, 0.04])
    cbar = plt.colorbar(contour_plot, orientation='horizontal',cax=cbar_ax)
    cbar.set_label(unitString,fontsize=11)

    axs.coastlines()

    axs.set_title(titleString)    
    
    return axs
    
```
