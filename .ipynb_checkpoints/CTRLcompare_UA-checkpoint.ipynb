## Comparison plots?

```python
def plotCtrlComparison(baseDS, compareDS, varSel,
                       baseStr,compareStr, pctDiff=False): 
    
    ## - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
    
    ## Get plotting limits (for raw values)
    minVal = np.nanmin([np.nanpercentile(baseDS[varSel].values, 5), 
                        np.nanpercentile(compareDS[varSel].values, 5)])
    maxVal = np.nanmin([np.nanpercentile(baseDS[varSel].values, 95), 
                        np.nanpercentile(compareDS[varSel].values, 95)])
    levsMap = np.arange(minVal,maxVal, (maxVal-minVal)/100.0)
    
    ## Get plotting limits (for difference plot)
    if pctDiff==True:
        diffVal = (((compareDS[varSel] - baseDS[varSel])/baseDS[varSel]).mean(dim='time').values)*100.0
    else:
        diffVal = (compareDS[varSel] - baseDS[varSel]).mean(dim='time').values
    minDiff = np.nanpercentile(diffVal,2) 
    maxDiff = np.nanpercentile(diffVal,98)
    cVal    = np.nanmax([np.abs(minDiff), np.abs(maxDiff)])
    levsDiff = np.arange(-cVal, cVal+((cVal*2)/100.0), (cVal*2)/100.0)
    
    ## - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
    
    ## Set up plot 
    fig,axs = plt.subplots(1,3, figsize=(15,10), subplot_kw={'projection':ccrs.PlateCarree()})
    axs = axs.ravel() 
    
    ## ctrlCase
    cplot = axs[0].contourf(baseDS.lon.values, baseDS.lat.values,
                    np.squeeze(baseDS[varSel].mean(dim='time').values), 
                    levsMap, cmap='jet', extend='both', transform=ccrs.PlateCarree())
    # Add coastline, set title
    axs[0].coastlines()
    axs[0].set_title(baseStr+': \n '+varSel+' (JJA)',fontsize=14)

    ## compareCase
    cplot = axs[1].contourf(compareDS.lon.values, compareDS.lat.values,
                    np.squeeze(compareDS[varSel].mean(dim='time').values), 
                    levsMap, cmap='jet', extend='both', transform=ccrs.PlateCarree())
    # Add coastline, set title
    axs[1].coastlines()
    axs[1].set_title(compareStr+': \n '+varSel+' (JJA)',fontsize=14)
    ## Add colorbar
    ax_position = axs[1].get_position()
    cbar_ax     = fig.add_axes([ax_position.x0-0.2, ax_position.y0-0.06, ax_position.width+0.1, 0.02])
    cbar        = plt.colorbar(cplot, orientation='horizontal',cax=cbar_ax)
    cbar.set_label(compareDS[varSel].units,fontsize=11)
    
    
    
    ## Plot diff 
    cplot = axs[2].contourf(compareDS.lon.values, compareDS.lat.values,
                    np.squeeze(diffVal), 
                    levsDiff, cmap='bwr', extend='both', transform=ccrs.PlateCarree())

    # Add coastline, set title
    axs[2].coastlines()
    axs[2].set_title(compareStr+' - '+baseStr+'\n'+varSel+' (JJA)',fontsize=14)

    ## Add colorbar
    ax_position = axs[2].get_position()
    cbar_ax     = fig.add_axes([ax_position.x0-0.02, ax_position.y0-0.06, ax_position.width+0.04, 0.02])
    cbar        = plt.colorbar(cplot, orientation='horizontal',cax=cbar_ax, format='%.2e')
    # cbar.set_label(L58fv_ctrlInst_noCold_h2['U10'].units,fontsize=11)
    if pctDiff==True:
        cbar.set_label('% Difference',fontsize=11)
    else: 
        cbar.set_label('Difference',fontsize=11)
    
    return
```
