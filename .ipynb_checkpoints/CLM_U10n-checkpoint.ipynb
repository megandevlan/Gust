```python
## Read in data 

scrtchDir = '/glade/scratch/mdfowler/archive/'
caseName  = 'FSCAM.T42_T42.TestARM97.58L_clm_computeU10n.002'

grdFile   = scrtchDir+caseName+'/lnd/hist/'+caseName+'.clm2.h0.1997-06-18-84585.nc' 
ptchFile  = scrtchDir+caseName+'/lnd/hist/'+caseName+'.clm2.h1.1997-06-18-84585.nc' 

gridDS    = xr.open_dataset(grdFile ,decode_times=True)
patchDS   = xr.open_dataset(ptchFile,decode_times=True)
```

```python
fig,axs = plt.subplots(1,1, figsize=(15,5))

axs.plot(gridDS.time.values, gridDS.U10.values, label='U10')
axs.plot(gridDS.time.values, gridDS.U10n.values, label='U10n')
axs.legend()
axs.set_title('U10 vs U10n \nComputed in CLM (ARM97)', fontsize=14)
axs.set_xlabel('time',fontsize=12)
axs.set_ylabel('m/s',fontsize=12)
```

```python
patchLabels = ['BareGround',            # PFT itype_veg = 0 \ itype_lunit = 1
               'Needleleaf Evergreen',  # PFT itype_veg = 1
               'Broadleaf Deciduous',   # PFT itype_veg = 7
               'C3 grass',              # PFT itype_veg = 13
               'C4 grass',              # PFT itype_veg = 14
               'C3 crop',               # PFT itype_veg = 15
               'C3 crop irrig',         # PFT itype_veg = 15
               'Urban hd',                 # itype_lunit = 8 (urban hd) 
               'Urban hd',                 # itype_lunit = 8 (urban hd) 
               'Urban hd',                 # itype_lunit = 8 (urban hd) 
               'Urban hd',                 # itype_lunit = 8 (urban hd) 
               'Urban hd',                 # itype_lunit = 8 (urban hd) 
               'Urban md',                 # itype_lunit = 9 (urban md) 
               'Urban md',                 # itype_lunit = 9 (urban md) 
               'Urban md',                 # itype_lunit = 9 (urban md) 
               'Urban md',                 # itype_lunit = 9 (urban md) 
               'Urban md',                 # itype_lunit = 9 (urban md) 
              ]
```

```python
diff    = patchDS.U10n.values - patchDS.U10.values

fig,axs = plt.subplots(1,1, figsize=(13,5))

axs.plot(gridDS.time.values, diff[:,0], color='firebrick', label=patchLabels[0])
axs.plot(gridDS.time.values, diff[:,1], color='orangered', label=patchLabels[1])
axs.plot(gridDS.time.values, diff[:,2], color='orange', label=patchLabels[2])
axs.plot(gridDS.time.values, diff[:,3], color='gold', label=patchLabels[3])
axs.plot(gridDS.time.values, diff[:,4], color='yellowgreen', label=patchLabels[4])
axs.plot(gridDS.time.values, diff[:,5], color='forestgreen', label=patchLabels[5])
axs.plot(gridDS.time.values, diff[:,6], color='dodgerblue', label=patchLabels[6])
axs.plot(gridDS.time.values, diff[:,16], color='darkviolet', label=patchLabels[16])

axs.legend(ncol=3)
axs.set_title('U10n - U10 (by patch)')
axs.axhline(0, color='grey', linestyle='--')
axs.set_ylim([-2.5,2.5])
```
